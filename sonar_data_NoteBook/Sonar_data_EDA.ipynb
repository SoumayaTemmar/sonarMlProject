## Visualization

```python
sonar_data.columns = [f"Feature_{i}" for i in range(60)] + ["Target"]
sonar_data.head()
```

```python
# select the most informative features to plot(visualize), we cannot plot all 60
class_mean = sonar_data.groupby("Target").mean().T
class_mean.head()
```

```python
# get features that differ the most between R and M â†’ potentially discriminative

top_features = (class_mean['M'] - class_mean['R']).abs().sort_values(ascending=False)
top_features.head(10)
```

```python
top_three = top_features.head(3).index

for feature in top_three:
   fig, axes = plt.subplots(1,2,figsize=(10,4))
   sns.histplot(data=sonar_data, x=feature, hue='Target', kde=True, ax=axes[0])
   axes[0].set_title(f'Histogram of {feature}')
    
   sns.boxplot(x='Target', y=feature, data=sonar_data, ax=axes[1])
   axes[1].set_title(f'Boxplot of {feature}')
   
   plt.show()
```

```python
# correlation test
plt.figure(figsize=(12, 10))
sns.heatmap(sonar_data.iloc[:, :-1].corr(), cmap='coolwarm', center=0)
plt.title("Feature Correlation Heatmap")
plt.show()
```
